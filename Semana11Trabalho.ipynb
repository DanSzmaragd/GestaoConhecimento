De acordo com os coeficientes, se removermos o feature que tem o menor índice, poderemos diminuir a margem de erro

```python
feature_cols = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'RM', 'AGE', 'DIS', 'RAD',
       'TAX', 'PTRATIO', 'B', 'LSTAT']
X = df[feature_cols]
y = df.Media
```

```python
linreg.fit(X_train, y_train)
y_pred = linreg.predict(X_test)
```

```python
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
